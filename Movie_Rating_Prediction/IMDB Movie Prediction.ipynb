```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


df_movie=pd.read_csv("IMDb Movies India.csv",encoding=("ISO-8859-1"),sep=",",engine='python')
df_movie.dropna(inplace=True)
df_movie.head()
```

```python
df_movie.describe()
```

```python
df_movie.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

df_movie['Genre']= labelencoder.fit_transform(df_movie['Genre'])

df_movie.head()
```

```python
df2=df_movie.drop(['Votes'],axis=1)
df2.head()
```

```python
df_final=df2.dropna()
```

```python
df_final.shape   
```

```python
df_final.head()   
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(df2['Rating'], bins=20, kde=True, color='red')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
plt.figure(figsize=(10, 6))
sns.countplot(y='Genre', data=df2, order=df2['Genre'].value_counts().index[:10], palette='viridis')
plt.title('Top 10 Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
plt.figure(figsize=(10, 6))
top_directors = df2['Director'].value_counts().nlargest(10)
sns.barplot(x=top_directors.values, y=top_directors.index, palette='rocket')
plt.title('Top 10 Directors')
plt.xlabel('Count')
plt.ylabel('Director')
plt.show()
```

```python

input=df_final.drop(['Duration','Name','Actor 1','Actor 2','Actor 3','Director','Year'], axis=1)
target=df_final['Rating']
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(input)
scaled_df = pd.DataFrame(scaled_data,columns=input.columns)
scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(input,target,test_size=0.35)
```

```python
print(Y_train)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
X_test = np.array(X_test)
```

```python
model.predict(X_test)
```

```python
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
r2_score = regressor.score(X_test,Y_test)
print(r2_score*100,'%')
```
